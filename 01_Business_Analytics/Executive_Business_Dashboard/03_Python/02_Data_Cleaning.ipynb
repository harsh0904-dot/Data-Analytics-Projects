# 02 Data Cleaning
Clean and validate NovaMart retail data.

```python
import pandas as pd
import numpy as np
```

```python
input_path = '../01_Dataset/Raw_Data/Dataset.xlsx'
output_path = '../01_Dataset/Clean_Data/NovaMart_Retail_Cleaned.csv'

df = pd.read_excel(input_path)
df.head()
```

```python
# Convert date columns (adjust names if needed)
for col in ['Order Date', 'Ship Date']:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')

df.dtypes
```

```python
missing_counts = df.isna().sum().sort_values(ascending=False)
duplicate_count = df.duplicated().sum()

print('Missing values by column:')
print(missing_counts[missing_counts > 0])
print('\nDuplicate rows:', duplicate_count)
```

```python
df = df.drop_duplicates().copy()

numeric_cols = [c for c in ['Sales', 'Profit', 'Discount', 'Quantity'] if c in df.columns]
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df[numeric_cols].describe()
```

```python
df.to_csv(output_path, index=False)
print('Cleaned dataset saved to:', output_path)
```
